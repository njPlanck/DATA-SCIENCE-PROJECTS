# Example 1 - A simple cost function

## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
#plot
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('x', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1), color = 'red')
plt.show()
```

## Slope and Derivatives

$f'(x) = df(x) = 2x + 1$

```python
def df(x):
    return 2*x + 1
```

```python
#plot function and derivative side by side
#1 Chart: Cost Function
plt.figure(figsize =[15,5])
plt.subplot(1,2,1)
plt.xlim([-3, 3])
plt.ylim([0,8])
plt.xlabel('x', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1), color = 'red')
#2 Chart: Derivative
plt.subplot(1,2,2)
plt.grid()
plt.xlim([-2, 3])
plt.ylim([-3,6])
plt.xlabel('x', fontsize = 16)
plt.ylabel('df(x)', fontsize = 16)
plt.plot(x_1, df(x_1), color = 'orange')
plt.show()
```

# Example 2 - Multiple Minima and Initial Guess & Advanced Functions

$$g(x) = x^4 - 4x^2 + 5$$

```python
#make more data
x_2 = np.linspace(-2,2,100)

def g(x):
    return x**4 - 4*x**2 + 5
    
def dg(x):
    return 4*x**3 - 8*x
```

```python
#plot function and derivative side by side
#1 Chart: Cost Function
plt.figure(figsize =[15,5])
plt.subplot(1,2,1)
plt.xlim([-2, 2])
plt.ylim([0.5,5.5])
plt.xlabel('x', fontsize = 16)
plt.ylabel('g(x)', fontsize = 16)
plt.plot(x_2, g(x_2), color = 'red')
#2 Chart: Derivative
plt.subplot(1,2,2)
plt.grid()
plt.xlim([-2, 2])
plt.ylim([-6,8])
plt.xlabel('x', fontsize = 16)
plt.ylabel('gf(x)', fontsize = 16)
plt.plot(x_2, dg(x_2), color = 'orange')
plt.show()
```

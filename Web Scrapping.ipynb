# Importing the important Libraries

```python
import requests #pip3 install requests
from bs4 import BeautifulSoup as bs #pip3 install beautifulsoup4
```

# Load the first page

```python
# loading the webpage with url 

r = requests.get("https://stackoverflow.com/questions/66583751/python-installing-requests-offline-as-whl-not-working")

# convert to a beuatiful soup object

soup = bs(r.content)

#print out soup

print(soup.prettify())
```

# Using BeautifulSoup to scrap the site

```python
first_header = soup.find("h2")
first_header
```

```python
#using soup.find() to find specific details on the site
headers = soup.find_all("h2")
headers
```

```python
# we can pass a list of what to find from the loaded webpage as below

first_header = soup.find_all(["h1","h2"])
first_header
```

```python
# passing attributes to the find/find_all function
paragraph = soup.find_all("p", attrs = {"class":"flex--item s-description"})
paragraph
```

```python
body = soup.find('body')
paragraph = body.find('p')
paragraph
#header = div.find('h1')
#header
```

```python
# importing re will healpp search for specific strings

import re
```

```python
paragraph = soup.find_all('p', string = re.compile('python'))
paragraph
```

# using Select function to extract details from the site

```python
paragraph = soup.select('p')
paragraph
```
